```python
import pandas as pd
import jieba
import jieba.analyse
#jieba.set_dictionary('dict.txt.big')
#jieba.load_userdict('dic.txt')
```

# 第一篇分析

```python
txt = (open('text.txt')).read().replace("\n","")
print("總字數：",len(txt))
wcut = ' '.join(jieba.cut(txt,cut_all=False))
print("總詞數：",len(wcut.split()))
wcut
```

```python
#word count
words={}
wd = wcut.split()
for w in wd:
    if len(w)<2:
        continue
    words[w] = words.get(w, 0) +1
dfw = pd.DataFrame(words,index=[0]).T
dfw = dfw.sort_values(by=0, ascending=False)

#keyword extract
prop = {}
for word, weight in jieba.analyse.extract_tags(txt, withWeight=True):
    prop[word] = round(weight, 3)

#keyword ranking
ranking = {}
for word, weight in jieba.analyse.textrank(txt, withWeight = True):
    ranking[word] = round(weight, 3)
#print(words)
print(prop)
print(ranking)
```

```python
df1 = pd.DataFrame(prop, index=[0])
df2 = pd.DataFrame(ranking, index=[0])
df1 = df1.T
df2 = df2.T
key1 = pd.concat([df1, dfw], axis=1, join='inner')
key2 = pd.concat([df2, dfw], axis=1, join='inner')
key1.columns=['詞比','詞數']
key2.columns=['詞分','詞數']
print(key1.head(10))
print(key2.head(10))
```

```python
#句子關鍵詞 textrank
pa = (open('text.txt')).read()
pa = pa.replace("\n\n","\n").replace('\n-','')

para = pa.split('\n')

rank_para = []
for i in para:
    ranking={}
    for word, weight in jieba.analyse.textrank(i, topK=5,withWeight = True):
        ranking[word] = round(weight, 3)
    rank_para.append(ranking)

rank_para
```

```python
import jieba.posseg as pseg
ww = {}
jieba.enable_paddle()
wa = pseg.cut(txt,use_paddle=True)
for word, flag in wa:
    ww[word]=flag
    print('%s %s' % (word, flag))
    
```

```python
"""n	普通名词	f	方位名词	s	处所名词	t	时间
nr	人名	ns	地名	nt	机构名	nw	作品名
nz	其他专名	v	普通动词	vd	动副词	vn	名动词
a	形容词	ad	副形词	an	名形词	d	副词
m	数量词	q	量词	r	代词	p	介词
c	连词	u	助词	xc	其他虚词	w	标点符号
PER	人名	LOC	地名	ORG	机构名	TIME	时间"""
speech = {}
for key, value in ww.items():
    speech[value] = speech.get(value,0) +1

sort_speech=[]    
for key, value in speech.items():
    sort_speech.append((value,key))
sorted(sort_speech,reverse=True)[:10]
```

### 第一篇情感分析

```python
from snownlp import SnowNLP
import re
sim = (open('sim.txt')).read()
sim = sim.replace("\n\n","\n").replace('\n-','')
s1 = re.split('。|！|？|\n',sim)
s1

emotion={}
em_avg = []
for i in s1:
    if i == '' or i =='」' or i =='-':
        continue
    s2 = SnowNLP(i)
    s3 = s2.sentiments
    emotion[i] = round(s3,3)
    em_avg.append(s3)
```

```python
print("文章平均情感分：",round(sum(em_avg)/len(em_avg),3))
dfe = pd.DataFrame(emotion, index=['情感分']).T
pd.options.display.max_rows = None
dfe
```

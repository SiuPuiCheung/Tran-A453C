```python
import pandas as pd
import jieba
import jieba.analyse
```

```python
txt = pd.read_csv("cn_corp.csv")
txt = txt[['title','content']]
txt
```

# 標題分析

```python
topic = txt['title'].astype(str)
topic = topic
topic
```

```python
topic2 = []
for i in topic:
    wcut = ' '.join(jieba.cut(i,cut_all=False))
    topic2.append(wcut)
topic2
```

```python
tw = []
tp = []
tr = []
for i in topic2:
    words={}
    wd = i.split()
    for w in wd:
        if len(w)<2:
            continue
        words[w] = words.get(w, 0) +1
    tw.append(words)

    #keyword extract
    prop = {}
    for word, weight in jieba.analyse.extract_tags(i, withWeight=True):
        prop[word] = round(weight, 3)
    tp.append(prop)

    #keyword ranking
    ranking = {}
    for word, weight in jieba.analyse.textrank(i, withWeight = True):
        ranking[word] = round(weight, 3)
    tr.append(ranking)
```

```python
#詞性分析
import jieba.posseg as pseg
ww = {}
jieba.enable_paddle()
wa = pseg.cut(topic[0],use_paddle=True)
for word, flag in wa:
    ww[word]=flag
```

```python
pos = []
for i in range(0,len(topic)):
    ww = {}
    jieba.enable_paddle()
    wa = pseg.cut(topic[i],use_paddle=True)
    for word, flag in wa:
        ww[word]=flag
    pos.append(ww)
    
tpos = []
for i in pos:
    for key, val in i.items():
        tpos.append(val)
```

```python
"""n	普通名词	f	方位名词	s	处所名词	t	时间
nr	人名	ns	地名	nt	机构名	nw	作品名
nz	其他专名	v	普通动词	vd	动副词	vn	名动词
a	形容词	ad	副形词	an	名形词	d	副词
m	数量词	q	量词	r	代词	p	介词
c	连词	u	助词	xc	其他虚词	w	标点符号
PER	人名	LOC	地名	ORG	机构名	TIME	时间"""
pos_count = {}
for i in tpos:
    pos_count[i] = pos_count.get(i, 0) +1
pcdf = pd.DataFrame(pos_count, index=[0]).T
pcdf = pcdf.sort_values(by=0, ascending=False)
pcdf['proportion'] = pcdf[0]/sum(pcdf[0])
pcdf.head()
```

# 句子分析

```python
content = txt['content'].astype(str)
content = content
```

```python
con2 = []
for i in content[:20]:
    cwcut = ' '.join(jieba.cut(i,cut_all=False))
    con2.append(cwcut)
con2
```

```python
ctw = []
ctp = []
ctr = []
for i in con2:
    words={}
    wd = i.split()
    for w in wd:
        if len(w)<2:
            continue
        words[w] = words.get(w, 0) +1
    ctw.append(words)

    #keyword extract
    prop = {}
    for word, weight in jieba.analyse.extract_tags(i, withWeight=True):
        prop[word] = round(weight, 3)
    ctp.append(prop)

    #keyword ranking
    ranking = {}
    for word, weight in jieba.analyse.textrank(i, withWeight = True):
        ranking[word] = round(weight, 3)
    ctr.append(ranking)
ctr
```

```python
#詞性分析
cww = {}
jieba.enable_paddle()
cwa = pseg.cut(content[0],use_paddle=True)
for word, flag in cwa:
    cww[word]=flag

cww
```

```python
cpos = []
for i in range(0,1000):
    ww = {}
    jieba.enable_paddle()
    wa = pseg.cut(content[i],use_paddle=True)
    for word, flag in wa:
        ww[word]=flag
    cpos.append(ww)
    
ctpos = []
for i in cpos:
    for key, val in i.items():
        ctpos.append(val)
```
